## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

A SQL database uses tables to store data, while a NoSQL uses a different method.  Neither SQL nor NoSQL is strictly better than the other, they are useful for different functions.  A NoSQL database is better for storing information that is not easily shaped for a SQL database such as text documents, or social media interactions.





## Question 2

Describe a situation where a NoSQL database might be a better choice than a relational database, and explain your reasoning.

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

Storing connections from social media would be a situation where it is better to use a NoSQL database.  In this instance, the number of connections between keys would be astronomical, requiring tons of extra work storing this information in tables.  Using a Graph Database would be much easier as each key would be represented as a node and each connection as an edge.  A Graph Database is set up to represent this sort of concept without any extra work, so it is a natural choice for this sort of information.


## Question 3

How does MongoDB work? How is it different from a relational database?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________

MongoDB stores information as documents.  Each document is comprised of key-value pairs with the information itself being saved as the value.  This is different than a relational database which stores information in tables instead.






## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________
Fault tolerance is resiliency to failing servers.  If an RDD makes a query over several servers and some of those servers are not functioning, the query will be returned incomplete or incorrect.  To an RDD, Fault tolerance is the process of storing a backup of each server on the other servers.  In this way, if one server goes down, there are other servers which can supply the same information.







## Question 5

What is MapReduce? How is it related to Hadoop?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________
MapReduce is the concept of each server running a query on itself, then combining all the results to get the complete query.  MapReduce is the backbone of Hadoop as it allows much faster queries and functions across multiple servers, allowing the storage of huge sets while retaining functionality..
